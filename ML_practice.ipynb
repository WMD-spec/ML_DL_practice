## 학습용 데이터 분할

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    iris['data'],  # 입력 2차원
    iris['target'], # 답 1차원
    test_size=0.3, 
    random_state=100
)
x_train.shape,x_test.shape, y_train.shape, y_test.shape
```

```python
x_train
```

## KNN 모델

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train, y_train) # 미리 분할해뒀던 훈련 데이터로 모델을 훈련시킴
# 훈련된 모델을 이용한 값 예측 출력
y_pred = model.predict(x_test) # 입력값은 반드시 2차원이어야 함. 
print("예측값 -", y_pred)
print("실제값 _", y_test)
# 모델의 정확도 계산
print( sum(y_pred == y_test)/ len(x_test)) # 예측값과 실제값이 같으면 True반환, sum은 True를 1로, False를 0으로 보기 때문에 이것을 전체 값으로 나누면 정확도가 나온다
print(model.score(x_test,y_test)) # 위와 같은 정확도를 보여주지만 더 쉽게 예측에 사용할 특징들과 기존의 답을 넘겨줘서 예측값과의 정확도 확인

print("훈련", model.score(x_train,y_train))
print("평가", model.score(x_test,y_test))
```

### for 문을 이용한 최적의 k 찾기 및 시각화

```python
# DataFrame으로 모델 평가하기
import seaborn as sns
iris = sns.load_dataset('iris')
iris
# 관례적으로 X 는 입력값, y = label 
X = iris.drop(columns = ['species'])
y = iris['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify= y)


k_range = range(1,11)
train_score=[]
test_score=[]
for k in k_range:
    model = KNeighborsClassifier(n_neighbors = k).fit(X_train, y_train)
    train_score.append(model.score(X_train,y_train))
    test_score.append(model.score(X_test,y_test))

print(train_score)
print(test_score)
```

```python
import matplotlib.pyplot as plt
_= plt.plot(k_range, train_score, label = 'Train Accuracy')
_= plt.plot(k_range, test_score, label = 'Test Accuracy')
_= plt.xticks(k_range)
_= plt.title('Find Best K-value in iris')
_= plt.legend()
_= plt.grid()
plt.style.use('ggplot')

plt.show()
```

## 타이타닉 데이터를 KNN의 성능그래프로 작성하세요.
###  -feature: 숫자칼럼 만 가능합니다.
###  -글자칼럼을 적용하려면 글자를 숫자로 변환하여 적용하세요.

```python
import seaborn as sns
ti = sns.load_dataset('titanic')
ti
```

```python
ti.isna().sum()
ti.drop(columns='deck',inplace=True)
```

```python
ti['age'].fillna(ti['age'].median(), inplace=True)
```

```python
ti['embarked'].value_counts()
```

```python
ti.drop(columns='embark_town', inplace=True)
ti['embarked'].fillna('S', inplace=True)
```

```python
ti.isna().sum()
```

```python
ti.drop(columns=['class','alive'], inplace=True)
```

```python
ti.dtypes
```

```python
def sex_convert(x):
    if x == 'male':
        return 0
    if x == 'female':
        return 1
def embarked_convert(x):
    if x == 'S':
        return 0
    if x == 'C':
        return 1
    if x == 'Q':
        return 2
def who_convert(x):
    if x == 'man':
        return 0
    if x == 'woman':
        return 1
    if x == 'child':
        return 2
def am_convert(x):
    if x == True:
        return 1
    if x == False:
        return 0
def alone_conver(x):
    if x == True:
        return 1
    if x == False:
        return 0

ti['sex'] = ti['sex'].apply(sex_convert)
ti['embarked'] = ti['embarked'].apply(embarked_convert)
ti['who'] = ti['who'].apply(who_convert)
ti['adult_male'] = ti['adult_male'].apply(am_convert)
ti['alone'] = ti['alone'].apply(alone_conver)
```

```python
ti['sex'] = ti['sex'].replace({'female':0, 'male':1})
# 이런 식으로 하면 더 편함
```

```python
X = ti.drop(columns = 'survived')
y = ti['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)

k_range = range(1,11)
train_score = []
test_score = []

for k in k_range:
    knn_c = KNeighborsClassifier(n_neighbors=k).fit(X_train,y_train)
    train_score.append(knn_c.score(X_train, y_train))
    test_score.append(knn_c.score(X_test, y_test))
    
_= plt.plot(k_range, train_score, label = 'Train Accuracy')
_= plt.plot(k_range, test_score, label = 'Test Accuracy')
_= plt.xticks(k_range)
_= plt.title('Find Best K-value in titanic')
_= plt.legend()
_= plt.grid()
plt.show()
```

```python
model = KNeighborsClassifier(n_neighbors=4)
model.fit(X_train, y_train) # 미리 분할해뒀던 훈련 데이터로 모델을 훈련시킴
# 훈련된 모델을 이용한 값 예측 출력

print("훈련", model.score(X_train,y_train))
print("평가", model.score(X_test,y_test))
```
