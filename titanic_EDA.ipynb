# 탐색적 자료 분석 (EDA: Exploratory Data Analysis)

```python
ti['survived'].value_counts()
```

```python
ti.groupby('embarked')['survived'].sum() # 항구별 생존자수
```

```python
ti.groupby('embarked')['survived'].mean() # 항구별 생존률
```

```python
ti.groupby('embarked')['survived'].count() # 항구별 탑승자수
```

```python
ti.groupby('embarked')['survived'].agg(['sum','count','mean'])
```

```python
ti.groupby('sex')['survived'].agg(['sum','count','mean'])
```

```python
ti.groupby('alone')['survived'].agg(['sum','count','mean'])
```

```python
ti.groupby('pclass')['survived'].agg(['sum','count','mean'])
```

```python
ti.groupby(['who','pclass'])['survived'].agg(['sum','count','mean'])
```

```python
ti.groupby(['alone','sex'])['survived'].agg(['sum','count','mean'])
```

```python
ti.pivot_table(index='sex', columns='pclass', values='survived', aggfunc = ['sum','count','mean'])
```

```python
result = ti.groupby(['who','pclass'])['survived'].agg(['sum','count','mean'])
result.reset_index()
result.sort_values(by='mean', ascending=False).reset_index(drop=True)
```

```python
rich = ti['fare'].quantile(0.9)
rich
```
