```python
import pandas as pd

data= pd.read_excel('./CustomerDataSet.xls')
data
```

```python
# 데이터 복사본
data2 = data.copy()
# 거리계산을 위해 정규화 전처리 실행
from sklearn.preprocessing import MinMaxScaler
data2[["ItemsBought","ItemsReturned"]] = MinMaxScaler().fit_transform(data2[["ItemsBought","ItemsReturned"]])
data2
```

```python
import matplotlib.pyplot as plt

# 군집분석
from sklearn.cluster import KMeans
# 여러가지 군집 갯수를 테스트
plt.figure(figsize=(10,8))
for k in range(2,8):
    model = KMeans(n_clusters=k)
    pred = model.fit_predict(data2[["ItemsBought","ItemsReturned"]])
    # 2행 3열 그래프
    plt.subplot(2,3,k-1) # 그래프 순번은 1부터 시작
    plt.tight_layout()
    plt.title(f"K = {k}")
    plt.xlabel("ItemsBought")
    plt.ylabel("ItemsReturned")
    plt.scatter(data['ItemsBought'], data['ItemsReturned'], c=pred)
```

```python
# 군집분석
from sklearn.cluster import KMeans
# 여러가지 군집 갯수를 테스트
model = KMeans(n_clusters=4)
pred = model.fit_predict(data2[["ItemsBought","ItemsReturned"]])
plt.xlabel("ItemsBought")
plt.ylabel("ItemsReturned")
plt.scatter(data['ItemsBought'], data['ItemsReturned'], c=pred)
    
# 각 데이터에 제품번호 출력
for index,c_id, bought,returned,zipcode,product in data.itertuples():
    plt.annotate(f"{pred[index]} : {product}",(bought,returned))
plt.show()
```
