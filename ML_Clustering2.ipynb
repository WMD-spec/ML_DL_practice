```python
import pandas as pd

data= pd.read_excel('./student_info.xls')
data
```

```python
# 데이터 복사본
data2 = data.copy()
# 거리계산을 위해 정규화 전처리 실행
from sklearn.preprocessing import MinMaxScaler
data2[["Mark","Attended"]] = MinMaxScaler().fit_transform(data2[["Mark","Attended"]])
data2
```

```python
data['Name'].nunique()
```

```python
data.groupby('Name').mean()
```

```python
import matplotlib.pyplot as plt

# 군집분석
from sklearn.cluster import KMeans
# 여러가지 군집 갯수를 테스트
plt.figure(figsize=(10,8))
for k in range(2,11):
    model = KMeans(n_clusters=k)
    pred = model.fit_predict(data2[["Attended", "Mark"]])
    # 2행 3열 그래프
    plt.subplot(3,3,k-1) # 그래프 순번은 1부터 시작
    plt.tight_layout()
    plt.title(f"K = {k}")
    plt.xlabel("Attended")
    plt.ylabel("Mark")
    plt.scatter(data['Attended'], data['Mark'], c=pred)
```

```python
# 군집분석
from sklearn.cluster import KMeans
# 여러가지 군집 갯수를 테스트
plt.figure(figsize=(10,10))
model = KMeans(n_clusters=3)
pred = model.fit_predict(data2[["Attended","Mark"]])
plt.xlabel("Attended")
plt.ylabel("Mark")
plt.scatter(data['Attended'], data['Mark'], c=pred)
    
# 각 데이터에 제품번호 출력
for index,Semester, Name,Course,Mark,Attended in data.itertuples():
    plt.annotate(f"{pred[index]} : {Name}",(Mark,Attended))
plt.show()
```

```python
data_grouped = data.groupby("Name").mean()
data_grouped
```

```python
# 거리계산을 위해 정규화 전처리 실행
from sklearn.preprocessing import MinMaxScaler
data_grouped[["Mark","Attended"]] = MinMaxScaler().fit_transform(data_grouped[["Mark","Attended"]])
data_grouped
```

```python
# 군집분석
from sklearn.cluster import KMeans
# 여러가지 군집 갯수를 테스트
plt.figure(figsize=(10,8))
for k in range(2,8):
    model = KMeans(n_clusters=k)
    pred = model.fit_predict(data_grouped[["Attended", "Mark"]])
    # 2행 3열 그래프
    plt.subplot(2,3,k-1) # 그래프 순번은 1부터 시작
    plt.tight_layout()
    plt.title(f"K = {k}")
    plt.xlabel("Attended")
    plt.ylabel("Mark")
    plt.scatter(data_grouped['Attended'], data_grouped['Mark'], c=pred)
```

```python
model = KMeans(n_clusters=3)
pred = model.fit_predict(data_grouped[["Attended","Mark"]])
# 2행 3열 그래프
plt.figure(figsize=(10,6)) #그래프 크기 지정
plt.subplot(2,3,k-1) #그래프 순번은 1부터 시작
plt.tight_layout()
plt.title(f"K={k}")
plt.xlabel("Attended")
plt.ylabel("Mark")
plt.scatter( 
    data_grouped["Attended"], 
    data_grouped["Mark"],
    c = pred # color : 같은 값끼리 같은 색을 가짐
) 
#각 데이터에 제품번호,지역번호 출력
for name, mark, attended in data_grouped.itertuples():
    plt.annotate( f"{name}",(attended+0.05,mark)  )
plt.show()
```
